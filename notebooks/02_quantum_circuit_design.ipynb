```python

from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes

# 1. The "Encoder" (Translates your 4 PCA numbers into the qubits)
feature_map = ZZFeatureMap(feature_dimension=4, reps=1, entanglement='linear')

# 2. The part the computer will learn to turn
ansatz = RealAmplitudes(num_qubits=4, reps=1)
#parameters = num_qubits x (reps + 1)

# 3. The Full Model
circuit = feature_map.compose(ansatz)

# 4. DRAW IT 
circuit.draw(output='mpl')
```

```python
# 4. DRAW IT 
circuit.draw(output='text')
```
